# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
# Importation du dataset
import numpy as np
import matplotlib.pyplot as plt
# Chargement du fichier CSV en sautant la première ligne (header)
data = np.genfromtxt('/content/Iris.csv', delimiter=',', skip_header=1)

print(f"Dataset chargé avec succès")
print(f"   Shape: {data.shape}")
print(f"   Type: {data.dtype}")
```

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
print("Séparation des features et des espèces")

# Slicing: extraire les colonnes 0 à 3 (features) et la colonne 4 (species)
features = data[:, :4]  # Toutes les lignes, colonnes 0 à 3
species = data[:, 4]    # Toutes les lignes, colonne 4

print(f"Features shape: {features.shape}")
print(f"Species shape: {species.shape}")
print(f"Premières valeurs de species: {species[:5]}")
```

# Partie C — Transformations Avancées avec NumPy

### C11. Espèces uniques
Listez les espèces uniques avec `np.unique` et comptez leurs occurrences.


```python
especes_uniques, counts = np.unique(species, return_counts=True)

print("Espèces uniques et leurs occurrences:")
for espece, count in zip(especes_uniques, counts):
    print(f"   Espèce {int(espece)}: {count} observations")
```

### C12. Reshape par espèce
Transformez `features` en un array de forme `(3, 50, 4)` nommé `features_by_species`.


```python
# Reshape en (3 espèces, 50 observations, 4 features)
features_by_species = features.reshape(3, 50, 4)

print(f"Shape après reshape: {features_by_species.shape}")
print(f"Dimension 0: {features_by_species.shape[0]} espèces")
print(f"Dimension 1: {features_by_species.shape[1]} observations par espèce")
print(f"Dimension 2: {features_by_species.shape[2]} features")
```

### C13. Concaténation
Concaténez horizontalement les moyennes par colonne avec un array de zéros.


```python
# Moyennes par colonne (array de shape (4,))
moyennes_col = np.mean(features, axis=0)

# Array de zéros de même shape
zeros = np.zeros(4)

# Concaténation horizontale
concat_result = np.concatenate([moyennes_col, zeros])

print(f"Moyennes: {moyennes_col}")
print(f"Zéros: {zeros}")
print(f"Résultat concaténé: {concat_result}")
print(f"Shape finale: {concat_result.shape}")
```

### C14. Data cleaning (simulation)
Simulez des valeurs NaN et remplacez-les avec `np.nan_to_num`.


```python
# Créer une copie pour simulation
features_nan = features[:5, :].copy()

# Introduire des NaN
features_nan[0, 0] = np.nan
features_nan[2, 1] = np.nan

print("Array avec NaN:")
print(features_nan)

# Remplacer les NaN par 0
features_cleaned = np.nan_to_num(features_nan)

print("\nArray après nettoyage:")
print(features_cleaned)
```

### C15. Broadcasting
Ajoutez 1 à toutes les valeurs de la première colonne de `features`.


```python
# Créer une copie pour ne pas modifier l'original
features_broadcast = features.copy()

# Ajouter 1 à toute la première colonne
features_broadcast[:, 0] = features_broadcast[:, 0] + 1

print(f"Avant (5 premières valeurs): {features[:5, 0]}")
print(f"Après (5 premières valeurs): {features_broadcast[:5, 0]}")
```

### C16. np.where
Créez un array binaire indiquant si la longueur des pétales est supérieure à 4.0.


```python
#Créer un array binaire: 1 si petal_length > 4.0, sinon 0
petal_length = features[:, 2]
binary_array = np.where(petal_length > 4.0, 1, 0)

print(f"Nombre de pétales > 4.0 cm: {np.sum(binary_array)}")
print(f"Nombre de pétales <= 4.0 cm: {np.sum(binary_array == 0)}")
print(f"Premières 10 valeurs: {binary_array[:10]}")
```

### E24. Bar chart avec error bars

```python
# Calculer moyennes et écarts-types par espèce
setosa_means = np.mean(features[:50], axis=0)
versicolor_means = np.mean(features[50:100], axis=0)
virginica_means = np.mean(features[100:150], axis=0)

setosa_std = np.std(features[:50], axis=0)
versicolor_std = np.std(features[50:100], axis=0)
virginica_std = np.std(features[100:150], axis=0)

x = np.arange(4)
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(x - width, setosa_means, width, yerr=setosa_std, label='Setosa',
       capsize=5, color='red', alpha=0.7)
ax.bar(x, versicolor_means, width, yerr=versicolor_std, label='Versicolor',
       capsize=5, color='green', alpha=0.7)
ax.bar(x + width, virginica_means, width, yerr=virginica_std, label='Virginica',
       capsize=5, color='blue', alpha=0.7)

ax.set_ylabel('Valeur moyenne (cm)', fontsize=12)
ax.set_title('Moyennes et écarts-types par espèce', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(['Sepal\nLength', 'Sepal\nWidth', 'Petal\nLength', 'Petal\nWidth'])
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Calculer la variance globale pour chaque espèce
setosa_var = np.var(features[:50])
versicolor_var = np.var(features[50:100])
virginica_var = np.var(features[100:150])

print(f"Variance globale par espèce:")
print(f"   Setosa:     {setosa_var:.4f}")
print(f"   Versicolor: {versicolor_var:.4f}")
print(f"   Virginica:  {virginica_var:.4f}")

# Identifier l'espèce avec la plus grande variance
variances = [setosa_var, versicolor_var, virginica_var]
espece_max_var = ['Setosa', 'Versicolor', 'Virginica'][np.argmax(variances)]

print(f"\n Espèce avec la plus grande variance: {espece_max_var}")

# Visualiser les distributions de cette espèce
if espece_max_var == 'Virginica':
    espece_data = features[100:150]
elif espece_max_var == 'Versicolor':
    espece_data = features[50:100]
else:
    espece_data = features[:50]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for i, ax in enumerate(axes.flat):
    ax.hist(espece_data[:, i], bins=10, color='darkblue', edgecolor='black', alpha=0.7)
    ax.set_xlabel(f'{feature_names_full[i]} (cm)', fontsize=10)
    ax.set_ylabel('Fréquence', fontsize=10)
    ax.set_title(f'{feature_names_full[i]} - {espece_max_var}', fontsize=12, fontweight='bold')
    ax.grid(axis='y', alpha=0.3)

plt.suptitle(f'Distributions pour {espece_max_var} (variance maximale)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
